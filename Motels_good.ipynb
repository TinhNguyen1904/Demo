```python
import bs4
import pandas as pd
import requests
```

```python
url = 'https://phongnhatro.com/cho-thue-phong-tro-nha-tro/can-tho/page/3'
def get_page_content(url):
   page = requests.get(url,headers={"Accept-Language":"en-US"})
   return bs4.BeautifulSoup(page.text,"html.parser")
soup = get_page_content(url)
```

```python
tennhatro = []
gia = []
diachi = []
quan = []
dientich = []
noidung = []
luotxem = []
```

```python
for mo in soup.findAll('div',class_='arttitle'):
  if ((not mo.findAll('span',class_='gia')) &
  (not mo.findAll('span',class_='diachi'))&
  (not mo.findAll('span',class_='dientich'))&
  (not mo.findAll('p',class_='tomtat'))&
  (not mo.findAll('span',class_='luotxem'))&
  (not mo.findAll('div',class_='qhlayout'))):
    continue
```

```python
tennhatro = [ce.text for ce in soup.findAll('div',class_='arttitle')]
gia = [ce.text for ce in soup.findAll('span',class_='gia')]
diachi = [rt.text for rt in soup.findAll('span',class_='diachi')]
quan = [rt.text for rt in soup.findAll('div',class_='qhlayout')]
dientich = [ce.text for ce in soup.findAll('span',class_='dientich')]
noidung = [rt.text for rt in soup.findAll('p',class_='tomtat')]
luotxem = [ce.text for ce in soup.findAll('span',class_='luotxem')]
```

```python
df={'tennhatro':tennhatro,  
    'gia':gia,
    'diachi':diachi,
    'quan':quan,
    'dientich':dientich,
     'noidung':noidung,
      'luotxem':luotxem}
danhsach = pd.DataFrame(df)
danhsach.to_csv('3.csv',encoding="utf-8-sig",index=False)
danhsach
```
